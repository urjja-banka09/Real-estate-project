```python
import requests
import pandas as pd

# Google Places API key
api_key = "YOUR_GOOGLE_PLACE_API_KEY"  # Replace with your actual API key

# Coordinates of Purva Westend, Block A
latitude = 12.902111609527427
longitude = 77.63996575110015
radius = 10000  # Radius in meters

# Define the search keyword
keyword = "apartment OR residential OR real estate"

# Construct the API request URL
url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={latitude},{longitude}&radius={radius}&keyword={keyword}&key={api_key}"

# Send the request to the Google Places API
response = requests.get(url)
data = response.json()

# Initialize an empty list to store the project data
projects = []

# Check if the request was successful
if data['status'] == 'OK':
    for place in data['results']:
        # Extract details from each place
        name = place['name']
        address = place.get('vicinity', 'N/A')
        place_id = place['place_id']  # Extract place_id
        lat = place['geometry']['location']['lat']
        lng = place['geometry']['location']['lng']

        # Append the data to the projects list
        projects.append({
            "name": name,
            "address": address,
            "place_id": place_id,
            "latitude": lat,
            "longitude": lng
        })
else:
    print("Error fetching data from Google Places API:", data['status'])

# Convert the list of projects to a DataFrame
df = pd.DataFrame(projects)

# Display the DataFrame
display(df)

# Save the DataFrame to an Excel file (optional)
df.to_excel('nearby_real_estate_projects.xlsx', index=False)
print("Project data has been saved to 'nearby_real_estate_projects.xlsx'")
```

```python
import numpy as np

# Function to calculate distance between two coordinates (Haversine formula)
def haversine(lon1, lat1, lon2, lat2):
    R = 6371  # Earth radius in kilometers
    dlon = np.radians(lon2 - lon1)
    dlat = np.radians(lat2 - lat1)
    a = np.sin(dlat/2) * np.sin(dlat/2) + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon/2) * np.sin(dlon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    distance = R * c
    return distance

# Initialize a dictionary to store amenities for each project
project_amenities = {}

# Iterate through each project
for _, project in merged_df.iterrows():
    project_name = project['name']
    project_lat = project['latitude']
    project_lon = project['longitude']

    # Calculate distance to each POI
    df_pois['distance_to_project'] = df_pois.apply(lambda poi: haversine(project_lon, project_lat, poi['longitude'], poi['latitude']), axis=1)

    # Filter POIs within the desired distance in kilometer(Km)
    nearby_pois = df_pois[df_pois['distance_to_project'] <= 2]

    # Summarize the amenities
    poi_summary = nearby_pois['poi_name'].value_counts().to_dict()

    # Store the summary in the dictionary
    project_amenities[project_name] = poi_summary

# Convert the dictionary to a DataFrame for easy comparison
amenities_df = pd.DataFrame(project_amenities).T.fillna(0)
# amenities_df.rename(columns={amenities_df.columns[0]: "name"}, inplace=True)

amenities_df.index.name = 'name'

amenities_df.to_excel('aminities.xlsx', index=False)
display(amenities_df)
```

```python
# Assign weights to POI types
poi_weights = {
    'Grocery Stores': 1.0,
    'Shopping Malls': 0.8,
    'Restaurants': 1.2,
    'Pharmacies': 1.0,
    'Public Transport Stops': 1.5,
    'Schools': 1.5,
    'Colleges': 1.7,
    'Daycare Centers': 1.2,
    'Hospitals': 2.0,
    'Clinics': 1.2,
    'Dentists': 1.1,
    'Parks': 1.0,
    'Gyms/Fitness Centers': 1.1,
    'Sports Complexes': 1.3,
    'Cinemas': 0.9,
    'Libraries': 1.0,
    'Banks/ATMs': 0.8,
    'Post Offices': 0.7,
    'Police Stations': 1.5,
    'Fire Stations': 1.4,
    'Churches': 0.8,
    'Temples': 0.8,
    'Mosques': 0.8,
    'Synagogues': 0.8,
    'Fuel Stations': 0.6,
    'Electric Vehicle Charging Stations': 0.6,
    'Parking Lots': 0.5,
    'Community Centers': 1.1,
    'Government Offices': 1.0,
    'Recycling Centers': 0.7,
    'Hotels': 0.9,
    'Guest Houses/B&Bs': 0.8,
    'Pet Services': 1.0,
    'Laundromats/Dry Cleaners': 0.6,
    'Hardware Stores': 0.7
}

# Initialize a list to store the proximity scores
proximity_scores = []

# Calculate proximity score for each project
for _, project in merged_df.iterrows():
    project_name = project['name']
    project_lat = project['latitude']
    project_lon = project['longitude']

    df_pois['distance_to_project'] = df_pois.apply(lambda poi: haversine(project_lon, project_lat, poi['longitude'], poi['latitude']), axis=1)

    # Filter POIs within 500 meters (0.5 km)
    nearby_pois = df_pois[df_pois['distance_to_project'] <= 2]

    # Calculate the proximity score
    score = 0
    for poi_name, count in nearby_pois['poi_name'].value_counts().items():
        score += poi_weights.get(poi_name, 0) * count

    # Append the score to the list
    proximity_scores.append({'name': project_name, 'proximity_score': score})

# Convert the list to a DataFrame
proximity_df = pd.DataFrame(proximity_scores)
display(proximity_df)
```

```python
import requests
import pandas as pd
import time

# Your Hugging Face API key
API_KEY = "YOUR_HUGGINGFACE_API_KEY"

# API URL for summarization
API_URL = "https://api-inference.huggingface.co/models/facebook/bart-large-cnn"

headers = {
    "Authorization": f"Bearer {API_KEY}"
}

# Function to call the Hugging Face API for summarization
def summarize_text(text, retries=3):
    payload = {
        "inputs": text,
        "parameters": {"max_length": 150, "min_length": 50, "do_sample": False},
    }
    for attempt in range(retries):
        response = requests.post(API_URL, headers=headers, json=payload)
        if response.status_code == 200:
            return response.json()[0]['summary_text']
        elif response.status_code == 503:
            print(f"Model is busy. Attempt {attempt + 1}/{retries}. Retrying in {2**attempt} seconds...")
            time.sleep(2 ** attempt)  # Exponential backoff
        else:
            print(f"Error: {response.status_code}, {response.text}")
            return "Error during summarization."
    return "Failed to get summary after multiple attempts."

# Function to generate a summary for each row in the DataFrame
def generate_summary(name, row):
    amenities_text = ""
    for column, value in row.items():
        if column != 'name' and pd.notna(value):
            amenities_text += f"{column}: {value}\n"

    if not amenities_text.strip():
        return "No amenities data available."

    prompt = f"Summarize the amenities for the real estate project '{name}':\n{amenities_text}"
    return summarize_text(prompt)

# Initialize an empty list to store summaries
summaries = []

# Process each row individually
for name, row in amenities_df.iterrows():
    summary = generate_summary(name, row)
    summaries.append(summary)
    # Add a delay to avoid rate limiting
    time.sleep(1)

# Add the summaries to the DataFrame
amenities_df['summary'] = summaries

# Display the DataFrame with summaries
display(amenities_df[['summary']])
```
